```python
data = {
    'name': ['Alice','Bob','Charlie','David','Emily'],
    'age': [25,np.nan,30,np.nan,35],
    'height':[165, 180, np.nan,175,np.nan],
    'weight':[55,70,60,np.nan,np.nan]
}
```

```python
df2 = pd.DataFrame(data)
df2['age'] = df2['age'].fillna(-1)
df2['age'] = df2['age'].astype(int)
df2['age'] = df2['age'].astype(str)
df2['age'] = df2['age'].replace('-1', 0)
df2['height'] = df2['height'].fillna(-1)
df2['height'] = df2['height'].astype(int)
df2['height'] = df2['height'].astype(str)
df2['height'] = df2['height'].replace('-1', 0)
df2['weight'] = df2['weight'].fillna(-1)
df2['weight'] = df2['weight'].astype(int)
df2['weight'] = df2['weight'].astype(str)
df2['weight'] = df2['weight'].replace('-1', 0)
df2
```

```python
df2 = df2.dropna(axis=1)
df2
```
