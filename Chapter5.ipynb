```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
def multiply(x):
  return x *1.1
```

```python
df['total_bill_multiply'] = df['total_bill'].apply(multiply)
```

```python
df
```

# Lamba #


```python
df['total_bill_labda'] = df['total_bill'].apply(lambda x: x *1.1)
```

```python
df
```

```python
df['total_bill_tips'] = df.apply(lambda row:row['total_bill'] + row['tip'], axis=1)
```

```python
df
```

```python
df['unit_price'] = df.apply(lambda row:row['total_bill'] / row['size'], axis=1)
```

```python
df
```
